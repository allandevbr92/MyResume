```python
import matplotlib.pyplot as plt

# Lista das colunas de notas
colunas_notas = ['NOTA_CN', 'NOTA_CH', 'NOTA_LC', 'NOTA_MT', 'NOTA_REDACAO']

# Dicionários para armazenar as medidas de cada coluna
maior_notas = {}
menor_notas = {}
medias = {}

for coluna in colunas_notas:
    # Maior nota
    maior_nota = df[coluna].max()
    maior_notas[coluna] = maior_nota

    # Menor nota
    menor_nota = df[coluna].min()
    menor_notas[coluna] = menor_nota

    # Média
    media = df[coluna].mean()
    medias[coluna] = media

# Gráfico de barras para a maior nota
plt.figure(figsize=(10, 6))
plt.bar(maior_notas.keys(), maior_notas.values())
plt.xlabel('Coluna')
plt.ylabel('Maior Nota')
plt.title('Maior Nota por Coluna')
plt.show()

# Gráfico de barras para a menor nota
plt.figure(figsize=(10, 6))
plt.bar(menor_notas.keys(), menor_notas.values())
plt.xlabel('Coluna')
plt.ylabel('Menor Nota')
plt.title('Menor Nota por Coluna')
plt.show()

# Gráfico de barras para a média
plt.figure(figsize=(10, 6))
plt.bar(medias.keys(), medias.values())
plt.xlabel('Coluna')
plt.ylabel('Média')
plt.title('Média por Coluna')
plt.show()

# Lista das colunas de notas
colunas_notas = ['NOTA_CN', 'NOTA_CH', 'NOTA_LC', 'NOTA_MT', 'NOTA_REDACAO']

for coluna in colunas_notas:
    # Maior nota
    maior_nota = df[coluna].max()
    print(f"Maior nota da coluna {coluna}: {maior_nota}")

    # Menor nota
    menor_nota = df[coluna].min()
    print(f"Menor nota da coluna {coluna}: {menor_nota}")

    # Média
    media = df[coluna].mean()
    print(f"Média da coluna {coluna}: {media}")

    # Moda
    moda = df[coluna].mode()
    print(f"Moda da coluna {coluna}: {moda[0]}\n")
```

```python
# Dicionário para armazenar as modas de cada coluna
modas = {}

for coluna in colunas_notas:
    # Moda
    moda = df[coluna].mode()
    modas[coluna] = moda[0]

# Gráfico de barras para a moda
plt.figure(figsize=(10, 6))
plt.bar(modas.keys(), modas.values())
plt.xlabel('Coluna')
plt.ylabel('Moda')
plt.title('Moda por Coluna')
plt.show()

```
